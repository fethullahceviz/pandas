#### 8.Merge the dataframes you created in questions 4-5-6-7. Expected output: example result


```python
columns     = pd.DataFrame(df.columns, columns=["columns"]) # columns dataframe
columns.reset_index(drop=True, inplace=True)
data_typ.reset_index(drop=True, inplace=True)
null.reset_index(drop=True, inplace=True)
nonNull.reset_index(drop=True, inplace=True)
unique.reset_index(drop=True, inplace=True)
data_frames = [columns, data_typ, null, nonNull, unique]
merge_df    = pd.concat(data_frames, join='outer', axis=1)
merge_df
```

### 11.Change the data type of "date" (object) column to datetime64 and reformat the date as DD/MM/YYYY.

```python
df.date = pd.to_datetime(df.date)
df.date = df.date.dt.strftime('%d/%m/%Y')
df.head(3)
```
